```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 


```

```python

#url="https://github.com/NF11/CESI-HumanFroYou/blob/main/data_set/employee_survey_data.csv"
#s=requests.get(url).content
#c=pd.read_csv(io.StringIO(s.decode('utf-8')))

df =pd.read_csv("data_set/employee_survey_data.csv")
```

```python
for col in df.select_dtypes('float64'):
    plt.figure()
    sns.distplot(df[col])
```

```python
ES = df["EnvironmentSatisfaction"].mean(axis=0)
JB = df["JobSatisfaction"].mean(axis=0)
WLB = df["WorkLifeBalance"].mean(axis=0)

ES_mean = ES #round(0) 
JB_mean = JB #round(0)
WLB_mean = WLB #round(0)

print("ES : ", ES_mean, " | JB : ", JB_mean,"| WLB : ", WLB_mean )

for a, value in df['EnvironmentSatisfaction'].isnull().iteritems():
    if value:
        df['EnvironmentSatisfaction'][a] = ES_mean

for a, value in df["JobSatisfaction"].isnull().iteritems():
    if value:
        df['JobSatisfaction'][a] = JB_mean

for a, value in df["WorkLifeBalance"].isnull().iteritems():
    if value:
        df['WorkLifeBalance'][a] = WLB_mean
```
